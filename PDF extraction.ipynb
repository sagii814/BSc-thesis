```python
import re
matches = re.findall(r'font-size:\S*pt',html_text)
```

```python
title_index = 0
for m in range(0,len(matches)):
    title = float(matches[0][10:-2])
    try:
        pt = float(matches[m][10:-2])
    except:
        pass
    if pt > title:
        title = pt
        title_index = m

print(matches[title_index])
```

```python
pattern = re.compile(matches[title_index]+'">\w')
pattern
```

```python
for match in re.finditer(pattern,html_text):
    print(html_text[match.start():match.end()+30])
    print('\n')
```

```python
from pdfminer.converter import PDFPageAggregator
from pdfminer.layout import LAParams, LTFigure, LTTextBox
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFPageInterpreter, PDFResourceManager
from pdfminer.pdfpage import PDFPage, PDFTextExtractionNotAllowed
from pdfminer.pdfparser import PDFParser
from pdfminer.pdfinterp import resolve1
```

```python
text = ""
stack = []

with open(unifile, 'rb') as f:
    parser = PDFParser(f)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = PDFPageAggregator(rsrcmgr, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    
    numPages = resolve1(doc.catalog['Pages'])['Count']
    print(f"Total number of pages: {numPages}")
    
    for p in range(0, numPages):
        page = list(PDFPage.create_pages(doc))[p]
        interpreter.process_page(page)
        layout = device.get_result()
    
        for obj in layout:
            if isinstance(obj, LTTextBox):
                text += obj.get_text()

            elif isinstance(obj, LTFigure):
                stack += list(obj)
```

```python
import pdfplumber
import re

pdf = pdfplumber.open(unifile)

for page in pdf.pages:
    print(page.extract_text())
    for pdf_table in page.extract_tables():
        table = []
        cells = []
        for row in pdf_table:
            if not any(row):
                # If all rows are empty, they are treated as the end of a record.
                if any(cells):
                    table.append(cells)
                    cells = []
            elif all(row):
                # If none of the rows is empty, then this article is new and the last one ends.
                if any(cells):
                    table.append(cells)
                    cells = []
                table.append(row)
            else:
                if len(cells) == 0:
                    cells = row
                else:
                    for i in range(len(row)):
                        if row[i] is not None:
                            cells[i] = row[i] if cells[i] is None else cells[i] + row[i]
        for row in table:
            print([re.sub('\s+', '', cell) if cell is not None else None for cell in row])
        print('---------- Dividing line ----------')

pdf.close()
```
