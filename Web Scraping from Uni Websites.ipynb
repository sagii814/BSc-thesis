```python
def entry_and_career(url):
    result = requests.get(url)
    soup = bs4.BeautifulSoup(result.text, 'lxml')

    entry_req = []
    tag = ''
    for i in range(0,len(soup.select('h2'))):
        if re.search(r'entry requirements', str(soup.select('h2')[i]), re.IGNORECASE) and tag != soup.select('h2')[i].parent:
            tag = soup.select('h2')[i].parent
            for s in tag.strings:
                t = ' '.join(s.split())
                if t != '':
                    entry_req.append(t)
    for i in range(0,len(soup.select('h3'))):
        if re.search(r'entry requirements', str(soup.select('h3')[i]), re.IGNORECASE) and tag != soup.select('h3')[i].parent:
            tag = soup.select('h3')[i].parent
            for s in tag.strings:
                t = ' '.join(s.split())
                if t != '':
                    entry_req.append(t)


    careers = []
    tag = ''
    for i in range(0,len(soup.select('h2'))):
        if re.search(r'careers', str(soup.select('h2')[i]), re.IGNORECASE) and tag != soup.select('h2')[i].parent:
            tag = soup.select('h2')[i].parent
            for s in tag.strings:
                t = ' '.join(s.split())
                if t != '':
                    careers.append(t)
    for i in range(0,len(soup.select('h3'))):
        if re.search(r'careers', str(soup.select('h3')[i]), re.IGNORECASE) and tag != soup.select('h3')[i].parent:
            tag = soup.select('h3')[i].parent
            for s in tag.strings:
                t = ' '.join(s.split())
                if t != '':
                    careers.append(t)
                    
    return [entry_req,careers]
```

```python
def combined_scraper(urls):
    scraped_courses = pd.DataFrame(columns = ['University', 'Course_name', 'Length', 'Credit', 'Language', 
                                              'Entry_requirements','Career_prospects'])
    
    for url in urls:
        try:
            result = requests.get(url)
            soup = bs4.BeautifulSoup(result.text, 'lxml')

            title = ''
            try:
                for j in re.split('\n|\xa0|\r|\t|-', soup.select('title')[0].get_text()):
                    if re.search(r'University',j):
                        title = j
            except:
                title = ''

            if len(soup.select('h1')) != 0:
                course_name = soup.select('h1')[0].get_text()
            elif len(soup.select('strong')) != 0:
                course_name = soup.select('strong')[0].get_text()
            else: course_name = ''
            course_name = ' '.join(course_name.split())


            text = []
            for s in soup.select('body')[0].strings:
                t = ' '.join(s.split())
                if t != '':
                    text.append(t)

            length = []
            for i in range(0, len(text)-1):
                match = re.search(r'(Duration:?|Course Length:?|Mode of \w+:?|No. of Semesters:?| Period of \w+:?)', text[i], re.IGNORECASE)
                if match != None:
                    length.append(text[i+1])
                    break
                match = re.search(r'(\d years?|\d semesters?)', text[i], re.IGNORECASE)
                if match != None:
                    length.append(match.group())

            credit = ''
            for m in range(0, len(text)-1):
                match = re.search(r'ECTS \w+:?|Credits?:?',text[m])
                if match != None and text[m+1].isnumeric():
                    credit = text[m+1]
                else:
                    match = re.search(r'\d{3,} ECTS|\d{3,} credits',text[m], re.IGNORECASE)
                    if match != None:
                        credit = match.group()
                        
                        
            lang = ''
            for i in range(0, len(text)-1):
                match = re.search(r'Language of',text[i], re.IGNORECASE)
                if match != None:
                    lang = text[i+1].split()[0]
                    
            
            entry_req = []
            tag = ''
            for i in range(0,len(soup.select('h2'))):
                if re.search(r'entry requirements', str(soup.select('h2')[i]), re.IGNORECASE) and tag != soup.select('h2')[i].parent:
                    tag = soup.select('h2')[i].parent
                    for s in tag.strings:
                        t = ' '.join(s.split())
                        if t != '':
                            entry_req.append(t)
            for i in range(0,len(soup.select('h3'))):
                if re.search(r'entry requirements', str(soup.select('h3')[i]), re.IGNORECASE) and tag != soup.select('h3')[i].parent:
                    tag = soup.select('h3')[i].parent
                    for s in tag.strings:
                        t = ' '.join(s.split())
                        if t != '':
                            entry_req.append(t)


            careers = []
            tag = ''
            for i in range(0,len(soup.select('h2'))):
                if re.search(r'careers', str(soup.select('h2')[i]), re.IGNORECASE) and tag != soup.select('h2')[i].parent:
                    tag = soup.select('h2')[i].parent
                    for s in tag.strings:
                        t = ' '.join(s.split())
                        if t != '':
                            careers.append(t)
            for i in range(0,len(soup.select('h3'))):
                if re.search(r'careers', str(soup.select('h3')[i]), re.IGNORECASE) and tag != soup.select('h3')[i].parent:
                    tag = soup.select('h3')[i].parent
                    for s in tag.strings:
                        t = ' '.join(s.split())
                        if t != '':
                            careers.append(t)
                            
            course = pd.Series([title, course_name, length,credit, lang, entry_req, careers],
                               index = ['University', 'Course_name', 'Length', 'Credit', 'Language','Entry_requirements','Career_prospects'])
            scraped_courses = scraped_courses.append(course, ignore_index= True)
            
        except:
            print(url)
        
    return scraped_courses
```
