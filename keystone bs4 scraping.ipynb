```python
import requests
import bs4
import pandas as pd

def scraper(base_url, page_num):
    #base format: 'https://www.bachelorstudies.com/Bachelor/Technology-Studies/?page={}'
    
    #initiate lists for scraped data
    course = []
    school = []
    location = []
    description = []
    degree_type = []
    pace = []
    duration = []
    language = []
    start_date = []
    mode = []
    link = []
    
    #iterate through all pages of the given url
    for p in range(1,page_num+1):
        scrape_url = base_url.format(p)
        res = requests.get(scrape_url)
        soup = bs4.BeautifulSoup(res.text, 'lxml')
        for i in soup.select('.program-listitem'):
            try:
                #must have info for each course
                course.append(i.select('h4')[0].getText())
                school.append(i.select('.school')[0].getText())
                location.append(i.select('.location')[0].getText().strip())
                description.append(i.select('.description')[0].getText().strip())
                degree_type.append(i.select('.degree')[0].select('.label-item')[0].getText().strip())
                link.append(i.select('a')[0].get('href'))
                
    
                try:
                    pace.append(i.select('.pace')[0].select('.label-item')[0].getText())
                except:
                    pace.append('Unknown')
                try:
                    duration.append(i.select('.duration')[0].select('.label-item')[0].getText())
                except:
                    duration.append('Unknown')
                try:
                    language.append(i.select('.languages')[0].select('.label-item')[0].getText().strip())
                except:
                    language.append('Unknown')
                try:
                    mode.append(i.select('.based')[0].select('.label-item')[0].getText())
                except:
                    mode.append('Unknown')
                try:
                    start_date.append(i.select('.start')[0].select('.label-item')[0].getText())
                except:
                    start_date.append('Unknown')
            except:
                print(scrape_url)
    
    #create dataframe from scraped data
    df = pd.DataFrame(list(zip(location, school, course, degree_type, pace, duration, language, mode, 
                               start_date, description, link)),
                       columns =['Location', 'School','Course', 'Degree_type','Pace','Duration', 'Language','Mode',
                                 'Start_date','Description', 'Link'])
    return df
```

```python
df_tech = scraper('https://www.bachelorstudies.com/Bachelor/Technology-Studies/?page={}', 111)
```
